```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

# **DEFINE FUNCTIONS**
#### USER SEARCH

```python
def user_search(contains_text = 'electronic groove'):
    
    base_search_url = 'https://api.mixcloud.com/search/'
    limit = 100
    search_params = {'type' : 'user', 'q' : contains_text, 'limit' : limit}
    
    search_results = json.loads(requests.get(url = base_search_url, params = search_params).content)
    
    results_list = [i['username'] for i in search_results['data']]      
    
    return sorted(results_list)
```

#### USER ANALYTICS

```python
def user_analytics(user = 'egpodcast'):
    
    base_url = 'https://api.mixcloud.com/'  # Mixcloud API URL
    
    preliminary_dict = json.loads(requests.get(base_url + user).content) # Create User Dictionary
    
    # Remove unneccessary information by creating a list of keys to remove -- then use dictionary comprehension to do so
    delete_keys = ['pictures', 'updated_time', 'is_premium', 'created_time', 'cover_pictures', 'key', 'picture_primary_color', 'is_pro']
    
    for key in delete_keys:
        if key in preliminary_dict:
            preliminary_dict.pop(key)
        else:
            print(f'Key {key} is not in user analytics dictionary')
    
    key_order = ['username', 'name', 'url', 'follower_count', 'cloudcast_count', 'following_count', 'favorite_count', 'listen_count', 'biog'] # 'city', 'country'
    
    user_analytics_dict = {key : preliminary_dict[key] for key in key_order}
    
    return user_analytics_dict
```

```python
user_search('masters')

query_df = pd.DataFrame()
for index, i in enumerate(user_search('masters')):
    df = pd.DataFrame(user_analytics(i), index=[index])
    query_df = pd.concat([query_df, df])
query_df
```

#### USER FOLLOWS

```python
def user_follows(user = 'egpodcast', follower_output = 50, results_per_page = 50):
    
    base_url = 'https://api.mixcloud.com/'
    following_path = '/following/?limit=' + str(results_per_page)
    
    following_number = json.loads(requests.get(base_url + user).content)['following_count']
    print(f'Showing {(follower_output // results_per_page + 1) * results_per_page} out of {following_number} followers for {user}')
    pages = follower_output // results_per_page + 1
    
    url_head = 'json.loads(requests.get('
    url_body = "'" + base_url + user + following_path + "')" + '.content)'
    url_tail = "['paging']['next']).content)"    
    
    user_following_dict = {}
    following_list = []
    for num in range(0, pages):
        for j in eval(url_head * (num + 1) + url_body + url_tail * num)['data']:
            following_list.append((j['url'], j['name']))
    user_following_dict[user] = sorted(following_list)

    return sorted(following_list)
```

#### MUSIC SEARCH

```python
def music_search(contains_text = 'darin epsilon', results_count = 100):
    
    url_search = 'https://api.mixcloud.com/search/'
    
    search_params = {'q' : contains_text, 'limit' : results_count, 'type' : 'cloudcast'}
    
    search_results = json.loads(requests.get(url = url_search, params = search_params).content)['data'] # paging or data

    search_keys = sorted(search_results[0].keys())
    search_keys.remove('user')
    search_keys.remove('slug')
    search_keys.remove('pictures')
    search_keys.remove('key')
    search_keys.remove('tags')

    results_list = []
    for i in range(len(search_results)):
        single_search_dict = {}
        for j in range(len(search_keys)):
            single_search_dict[str(search_keys[j])] = search_results[i][search_keys[j]]
        tags_list = []
        for h in search_results[i]['tags']:
            tags_list.append(h['url'])
        single_search_dict['tags'] = tags_list
        results_list.append(single_search_dict)
    
    results_df = pd.DataFrame(results_list)
    results_df['user'] = results_df['url'].apply(lambda x: x.replace('/', ' ').split()[2])
    results_df['playback_time_minutes'] = results_df['audio_length'].apply(lambda x: round(x / 60, 2))
    results_df = results_df[['user', 'name', 'listener_count', 'play_count', 'repost_count', 'favorite_count', 'comment_count', 'playback_time_minutes', 'created_time', 'updated_time', 'url', 'tags']]

    return results_df
```
